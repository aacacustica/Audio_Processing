```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold

df = pd.read_csv(r"D:\AUDIOSET_STRONG\audioset_66_classes_folds_trimmed.csv") 

# remove columns
df = df.drop(columns=['Unnamed: 0', 'fold'])

df
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

df['fold'] = 0  

for fold, (train_index, val_index) in enumerate(skf.split(X=df, y=df['category'])):
    df.loc[val_index, 'fold'] = fold

print(df['fold'].value_counts())
df = df [["filename", "fold", "category", "mid", "duration"]]


df.to_csv(r"D:\AUDIOSET_STRONG\audioset_66_classes_folds_trimmed_crossval.csv", index=False)
```
