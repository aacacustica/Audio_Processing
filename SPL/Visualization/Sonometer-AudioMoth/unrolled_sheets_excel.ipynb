```python
import pandas as pd
import os
```

```python
excel_path = r"\\192.168.205.117\AAC_Server\OCIO\OCIO_BILBAO\CAMPAÑA_4\3-Medidas\sonometro_bilbao\SONOMETRO\Datos_Balcon_de_la_Lola.xlsx"
# excel_file = pd.read_excel(excel_path, sheet_name=None, header=4)
excel_file = pd.read_excel(excel_path, sheet_name=None)
output_dir = os.path.dirname(excel_path)

print(output_dir)
output_dir = output_dir.split("\\")[:-2]
print(output_dir)
output_dir = "\\".join(output_dir)
print(output_dir)
# make sure the output directory exists
if not os.path.exists(output_dir):
    print(f"Created {output_dir} directory")
```

```python
# list the sheet n
excel_sheets = list(excel_file.keys())
excel_sheets_count = len(excel_sheets)
print(f"Excel file has {excel_sheets_count} sheets \nwhich are {excel_sheets}\n")

for sheet in excel_sheets:
    excel_file[sheet] = excel_file[sheet][["Date", "Value"]]
    # rename Date to datetime
    excel_file[sheet] = excel_file[sheet].rename(columns={"Date":"datetime"})
    # convert to datetime
    excel_file[sheet]["datetime"] = pd.to_datetime(excel_file[sheet]["datetime"], dayfirst=True)
    # order by datetime
    excel_file[sheet] = excel_file[sheet].sort_values(by="datetime")
    # remove nan values
    excel_file[sheet] = excel_file[sheet].dropna()
    print(excel_file[sheet].head())

    # save each sheet as a csv fil
    output_file = f"{sheet}/SONOMETRO/"
    if not os.path.exists(os.path.join(output_dir, output_file)):
        os.makedirs(os.path.join(output_dir, output_file))
        print(f"Created {os.path.join(output_dir, output_file)} directory")

    # save to csv
    excel_file[sheet].to_csv(os.path.join(output_dir, output_file, f"{sheet}.csv"), index=False)
```
